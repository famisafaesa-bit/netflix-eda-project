# Exploratory Data Analysis (EDA) on Netflix Dataset

 The dataset was collected from kaggle and contains information about Netflix Movies and TV

# Dataset Description

This dataset contains information about Netflix Movies and TV Shows including:

- Title
- Type
- Director
- Cast
- Country
- Release Year
- Rating
- Duration
- Genre
- Date Added

# Business Understanding

The goal of this EDA project is to analyze Netflix content and discover:

- Content distribution
- Popular genres
- Country-wise production
- Rating patterns
- Trends over years
- Movie vs TV Show analysis
